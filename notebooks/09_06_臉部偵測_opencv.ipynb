# 臉部偵測(Face Detection)

## 載入相關套件

```python
# 載入相關套件
import cv2
from cv2 import CascadeClassifier
from cv2 import rectangle
import matplotlib.pyplot as plt
from cv2 import imread
```

## 載入級聯分類器(face cascade file)

```python
# 載入臉部級聯分類器(face cascade file)
face_cascade = './haarcascade_frontalface_alt.xml'
classifier = cv2.CascadeClassifier(face_cascade)
```

## 載入圖檔

```python
# 載入圖檔
image_file = "./52-0.jpg"
image = imread(image_file)

# OpenCV 預設為 BGR 色系，轉為 RGB 色系
im_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 顯示圖像
plt.imshow(im_rgb)
plt.axis('off')
plt.show()
```

## 偵測臉部並顯示圖像

```python
# 偵測臉部
bboxes = classifier.detectMultiScale(image)
# 臉部加框
for box in bboxes:
    # 取得框的座標及寬高
    x, y, width, height = box
    x2, y2 = x + width, y + height
    # 加白色框
    rectangle(im_rgb, (x, y), (x2, y2), (255,255,255), 2)

# 顯示圖像
plt.imshow(im_rgb)
plt.axis('off')
plt.show()
```

## 載入另一圖檔

## 載入眼睛及微笑級聯分類器

```python
# 載入眼睛級聯分類器(eye cascade file)
eye_cascade = './haarcascade_eye_tree_eyeglasses.xml'
classifier = cv2.CascadeClassifier(eye_cascade)

# 載入微笑級聯分類器(smile cascade file)
smile_cascade = './haarcascade_smile.xml'
smile_classifier = cv2.CascadeClassifier(smile_cascade)
```

## 偵測臉部並顯示圖像

```python
# 偵測臉部
bboxes = classifier.detectMultiScale(image)
# 臉部加框
for box in bboxes:
    # 取得框的座標及寬高
    x, y, width, height = box
    x2, y2 = x + width, y + height
    # 加白色框
    rectangle(im_rgb, (x, y), (x2, y2), (255,0,0), 5)

# 偵測微笑
# scaleFactor=2.5：掃描時每次縮減掃描視窗的尺寸比例。
# minNeighbors=20：每一個被選中的視窗至少要有鄰近且合格的視窗數
bboxes = smile_classifier.detectMultiScale(image, 2.5, 20)
#微笑加框
for box in bboxes:
    # 取得框的座標及寬高
    x, y, width, height = box
    x2, y2 = x + width, y + height
    # 加白色框
    rectangle(im_rgb, (x, y), (x2, y2), (255,0,0), 5)
#     break

# 顯示圖像
plt.imshow(im_rgb)
plt.axis('off')
plt.show()
```
