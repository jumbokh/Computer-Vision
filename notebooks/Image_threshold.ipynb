```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('gradient.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY); # 
ret, output1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)     #  127  255 0
ret, output2 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV) #  127  0 255
ret, output3 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TRUNC)      #  127  127
ret, output4 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO)     #  127  0
ret, output5 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO_INV) #  127  0

cv2_imshow(img)
cv2_imshow(output1)
cv2_imshow(output2)
cv2_imshow(output3)
cv2_imshow(output4)
cv2_imshow(output5)
cv2.waitKey(0)    # 
```
